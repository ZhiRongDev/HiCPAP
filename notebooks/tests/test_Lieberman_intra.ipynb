```python
import numpy as np
import pandas as pd
from copy import deepcopy
np.set_printoptions(suppress=True)

# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

path = "/home/jordan990301/Projects/HiC-PC1_Approximation/data/Lieberman_Aiden-datasets/heatmaps"
```

```python
oe = pd.read_table(f"{path}/HIC_gm06690_chr10_chr10_1000000_obsexp.txt", index_col=0, header=1, sep="\t")
oe.pop(oe.columns[-1])

oe_np = oe.values
# remove 0 rows and 0 columns
# https://stackoverflow.com/questions/11188364/remove-zero-lines-2-d-numpy-array
oe_np = oe_np[~np.all(oe_np == 0, axis=1)]
oe_np = oe_np[:, ~np.all(oe_np == 0, axis=0)]

# zero means
oe_np = oe_np - oe_np.mean(axis=1, keepdims=True)

n = len(oe_np)
# y_np = oe_np.T / np.sqrt(n)
y_np = oe_np.T / np.sqrt(n - 1)
cov_oe_np = np.matmul(y_np.T, y_np)

# print(cov_oe_np, '\n')
# print(np.cov(oe_np))

corr_oe_np = np.empty(shape=(n, n))
corr_oe_np.fill(0)

for i in range(n):
    for j in range(n):
        corr_oe_np[i][j] = cov_oe_np[i][j] / np.sqrt(cov_oe_np[i][i] * cov_oe_np[j][j])

display(corr_oe_np)
np.corrcoef(oe_np)
```
