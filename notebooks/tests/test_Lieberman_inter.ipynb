```python
import numpy as np
import pandas as pd
from copy import deepcopy
np.set_printoptions(suppress=True)

# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

path = "/home/jordan990301/Projects/HiC-PC1_Approximation/data/Lieberman_Aiden-datasets/heatmaps"
```

```python
oe_all = None
chrom_list = ['1', '2', '3']

for i in chrom_list:
    oe_row = pd.read_table(f"{path}/HIC_gm06690_chr{i}_chr1_1000000_obsexp.txt", index_col=0, header=1, sep="\t")
    oe_row.pop(oe_row.columns[-1])
    for j in chrom_list:
        if j == '1':
            continue
        else:
            oe_element = pd.read_table(f"{path}/HIC_gm06690_chr{i}_chr{j}_1000000_obsexp.txt", index_col=0, header=1, sep="\t")
            oe_element.pop(oe_element.columns[-1])
            oe_row = pd.concat([oe_row, oe_element], axis=1)

    oe_all = pd.concat([oe_all, oe_row])

display(oe_all)
display(oe_row)

# # remove 0 rows and 0 columns
# # https://stackoverflow.com/questions/11188364/remove-zero-lines-2-d-numpy-array
# oe_np = oe_np[~np.all(oe_np == 0, axis=1)]
# oe_np = oe_np[:, ~np.all(oe_np == 0, axis=0)]

# # zero means
# # oe_np = oe_np - oe_np.mean(axis=1, keepdims=True)
```
